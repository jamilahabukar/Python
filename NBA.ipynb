# Task - Merge College info with NBA Player info

```python
file_path = "Players_College.csv"
players_college = pd.read_csv(file_path)

print("CSV file has been succefully read in")

players_college.head()
```

```python
nba_players = pd.merge(merged_df, players_college, on= 'Player', how='left')

nba_players.head()
```

# Calculate per game stats

```python
# Calculate PTS per game
nba_players['PTS per game'] = nba_players['PTS'] / nba_players['Games Played']

# Calculate TRB per game 
nba_players['TRB per game'] = nba_players['TRB'] / nba_players['Games Played']

# Calculate AST per game
nba_players['AST per game'] = nba_players['AST'] / nba_players['Games Played']

# Calculate TOV per game 
nba_players['TOV per game'] = nba_players['TOV'] / nba_players['Games Played']

# Calculate 3P per game 
nba_players['3P per game'] = nba_players['3P'] / nba_players['Games Played']

nba_players.head()
```

# Round up the Calculations

```python
# Calculate PTS per game
nba_players['PTS per game'] = round(nba_players['PTS'] / nba_players['Games Played'], 1)

# Calculate TRB per game 
nba_players['TRB per game'] = round(nba_players['TRB'] / nba_players['Games Played'], 1)

# Calculate AST per game
nba_players['AST per game'] = round(nba_players['AST'] / nba_players['Games Played'], 1)

# Calculate TOV per game 
nba_players['TOV per game'] = round(nba_players['TOV'] / nba_players['Games Played'], 1)

# Calculate 3P per game 
nba_players['3P per game'] = round(nba_players['3P'] / nba_players['Games Played'], 1)

nba_players.head()
```

# What players have played the most Games

```python
top_players = nba_players.nlargest(10, 'Games Played')


# create a bar plot 
plt.figure(figsize=(10, 6))
plt.bar(top_players['Player'], top_players['Games Played'], color='#647E05')
plt.title('Top 10 Players with the most Games Played')
plt.xlabel('Player')
plt.ylabel('Games Played')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Games Played vs. Points

```python
plt.figure(figsize=(10, 6))
plt.scatter(nba_players['Games Played'], nba_players['PTS'], color='#C9F919')
plt.title('Games played vs. Points (PTS)')
plt.xlabel('Games Played')
plt.ylabel('Points(PTS)')
plt.grid(True)
plt.show()
```

# Which college has produced the best Player (Points Per Game)

```python
# Group the data by college and calculate the average PTS per game for each college 
college_pts_avg = nba_players.groupby('College')['PTS per game'].mean().sort_values(ascending=False)

# Select the top 10 colleges with the highest average PTS per game 
top_colleges = college_pts_avg.head(10)

# Create a bar plot
plt.figure(figsize=(10, 6))
top_colleges.plot(kind='bar', color='Blue')
plt.title('Top 10 Colleges by Average Point per game (PTS per game)')
plt.xlabel('College')
plt.ylabel('Average Point per game (PTS per game)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
